## Load the dataset

Load the data available at dataset_url into a dataframe.


```python
# your code goes here

df_url = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")
```

## Find out the number of rows and columns


Start by exploring the numbers of rows and columns of data in the dataset.


Print the number of rows in the dataset.


```python
# your code goes here

no_rows = df_url.shape[0]
print('total_no_of_rows:', no_rows)
```

Print the number of columns in the dataset.


```python
# your code goes here

no_col = df_url.shape[1]
print('total_no_of_col:', no_col)
```

## Identify the data types of each column


Explore the dataset and identify the data types of each column.


Print the datatype of all columns.


```python
# your code goes here

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print("the datatypes of all columns are as follows : \n \n", df_url.dtypes)
```

Print the mean age of the survey participants.


```python
# your code goes here

df_url_mean = df_url['Age'].mean()
print("the mean age of the survey participants : ", df_url_mean)
```

The dataset is the result of a world wide survey. Print how many unique countries are there in the Country column.


```python
# your code goes here

df_url_unique = df_url['Country'].nunique()
print("the total no. of unique countries are :", df_url_unique)
```
